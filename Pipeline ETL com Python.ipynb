```python
# Criar o modelo
model = genai.GenerativeModel('gemini-2.5-flash')

def generate_ai_news(user):
    prompt = f"""Você é um especialista em Marketing Digital Bancário.
    
Crie uma mensagem para {user['name']} sobre a importância dos investimentos.
A mensagem deve ter no máximo 100 caracteres.
Não mostre mensagens sobre a quantidade de caracteres utilizados (exemplo 83 caracteres)."""
    
    response = model.generate_content(prompt)
    return response.text.strip('*')

for user in users:
    news = generate_ai_news(user)
    print(news)
    user['news'].append({
        "description": news
    })
```

```python
# Atualizar a lista de "news" de cada usuário na API com a nova mensagem gerada pela IA
def update_user(user):
    """Simula a atualização salvando no arquivo JSON local"""
    try:
        # Ler o arquivo JSON atual
        with open('users.json', 'r', encoding='utf-8') as file:
            all_users = json.load(file)
        
        # Encontrar e atualizar o usuário específico
        for i, usuario in enumerate(all_users):
            if usuario['id'] == user['id']:
                all_users[i] = user
                break
        
        # Salvar de volta no arquivo
        with open('users.json', 'w', encoding='utf-8') as file:
            json.dump(all_users, file, indent=2, ensure_ascii=False)
        
        return True
    except Exception as e:
        print(f"Erro ao atualizar usuário {user['id']}: {e}")
        return False

# Atualizar cada usuário
for user in users:
    success = update_user(user)
    if success:
        print(f"✅ Usuário {user['name']} atualizado com sucesso!")
    else:
        print(f"❌ Falha ao atualizar usuário {user['name']}")
```
