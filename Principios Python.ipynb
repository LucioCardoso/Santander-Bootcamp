#### 1.5.3. Função Range
Range é uma função built-in do Python, ela é usada para produzir uma sequência de números inteiros a partir de um início (inclusivo) para um fim (exclusivo).  Se usarmos range(i, j), será produzido: i, i+1, i+2, i+3, ..., j-1.

Ela recebe 3 argumentos: start (opcional), stop (obrigatório) e step (opcional).

```python
# range(stop) -> range object
# range(start, stop[, step]) -> range object

print(f"list(range(4)) = {list(range(4))}")

print(f"list(range(4, 10)) = {list(range(4, 10))}")

print(f"list(range(4, 10, 2)) = {list(range(4, 10, 2))}")
```

#### 1.5.4. While

```python
opcao = -1

while opcao != 0:
    opcao = int(input("[1] Sacar \n[2] Extrato \n[0] Sair \n: "))

    if opcao == 1:
        print("Sacando ...")
    elif opcao == 2:
        print("Exibindo o extrato ...")
else:
    print("Operação terminada!")
```

### 1.6.5. Fatiamento de String

Fatiamento de strings é uma técnica utilizada para retornar substrings (partes da string original), informando inicio (start), fim (stop) e passo (step): *[start: stop[, step]]*.

```python
# Fatiamento de Strings
nome = "Lúcio Fernando Diniz Cardoso"

print(f"nome: {nome}\n")
print(f"nome[0]: {nome[0]}\n")
print(f"nome[:5]: {nome[:5]}\n")
print(f"nome[6:]: {nome[6:]}\n")
print(f"nome[6:14]: {nome[6:14]}\n")
print(f"nome[6:14:2]: {nome[6:14:2]}\n")
print(f"nome[:]: {nome[:]}\n")
print(f"nome[::-1]: {nome[::-1]}")
```

## 1.7. Listas em Python
Listas em Python podem armazenar de maneira sequencial qualquer tipo de objeto.  Podemos criar listas utilizando o construtor **list**, a função range ou colocando valores separados por vírgula dentro de colchetes.  Listas são objetos mutáveis, portanto podemos alterar seus valores após a criação.

```python
# Listas
frutas = ["laranja", "maçã", "uva"]
frutas_vazias = []
letras = list("python")
numeros = list(range(10))
carro = ["Ferrari", "F8", 4200000, 2020, 2900, "São Paulo", True]

print(f"frutas = {frutas}\n")
print(f"frutas_vazias = {frutas_vazias}\n")
print(f"letras = {letras}\n")
print(f"numeros = {numeros}\n")
print(f"carro = {carro}\n")
```

### 1.7.1. Acesso direto
A lista é uma sequência, portanto podemos acessar seus dados utilizando índices.  Contamos o índice de determinada sequência a partir do zero.

```python
# Acesso direto à uma lista
frutas = ["laranja", "maçã", "uva", "pera"]

print(f"frutas[0] = {frutas[0]}\n")
print(f"frutas[2] = {frutas[2]}\n")
```

#### 1.7.1.1. Índices negativos

Sequências suportam indexação negativa.  A contagem começa em -1.

```python
# índice negativo
frutas = ["laranja", "maçã", "uva", "pera"]

print(f"frutas[-1] = {frutas[-1]}\n")
print(f"frutas[-3] = {frutas[-3]}")
```

### 1.7.2. Listas Aninhadas
Listas podem armazenar todos os tipos de objetos Python, portanto podemos ter listas que armazenam outras listas. Com isso, podemos criar estruturas bidimensionais (tabelas), e acessar informando os índices de linha e coluna.

```python
# Listas Aninhadas
matriz = [
    [1, "a", 2],
    ["b", 3, 4],
    [5, 6, "c"]
]

print(f"Matriz = {matriz}\n")
print(f"matriz[0] = {matriz[0]}\n")
print(f"matriz[0][0] = {matriz[0][0]}\n")
print(f"matriz[0][-1] = {matriz[0][-1]}\n")
print(f"matriz[-1][-1] = {matriz[-1][-1]}")
```

### 1.7.3. Fatiamento de Listas
Além de acessar elementos diretamente, podemos extrair um conjunto de valores de uma sequência.  Para isso basta passar o índice inicial e/ou final para acessar o conjunto.  Podemos ainda informar quantas posições o cursor deve "pular" no acesso.

```python
# Fatiamento de Listas
lista = ["p", "y", "t", "h", "o", "n"]

print(f"lista = {lista}\n")
print(f"lista[2:] = {lista[2:]}\n")
print(f"lista[:2] = {lista[:2]}\n")
print(f"lista[1:3] = {lista[1:3]}\n")
print(f"lista[0:3:2] = {lista[0:3:2] }\n")
print(f"lista[::] = {lista[::]}\n")
print(f"lista[::-1] = {lista[::-1]}")
```

```python
# Fatiando usando for
carros = ["gol", "celta", "palio", "uno"]
print(f"Lista carros = {carros}\n")

for carro in carros:
    print(carro)
```

### 1.7.4. Função enumerate
Às vezes é necessário saber qual o índice do objeto dentro do laço **for**.  Para isso, podemos usar a função **enumerate**.

```python
# Função enumerate
carros = ["gol", "celta", "palio", "uno"]
print(f"Lista carros = {carros}\n")

for indice, carro in enumerate(carros):
    print(f"Índice:{indice} da lista carros = {carro}")
```

### 1.7.5. Compreensão de listas
A compreensão de lista oferece uma sintaxe mais curta quando você deseja: criar uma nova lista com base nos valores de uma lista existente (filtro) ou gerar uma nova lista aplicando alguma modificação nos elementos de uma lista existente.

```python
# Compreensão de listas - Filtro versão 1
numeros = [1, 30, 21, 2, 9, 65, 34]
pares = []

print(f"Lista numeros = {numeros}")
print(f"Lista pares = {pares}\n")

for numero in numeros:
    if numero % 2 == 0:
        pares.append(numero)

print(f"Lista pares filtrados = {pares}")
```

```python
# Compreensão de listas - Filtro versão 2
numeros = [1, 30, 21, 2, 9, 65, 34]
pares = []

print(f"Lista numeros = {numeros}")
print(f"Lista pares = {pares}\n")

# retorno obrigatório, iteração for, filtro com if 
pares = [numero for numero in numeros if numero % 2 == 0]
print(f"Lista pares filtrados = {pares}")
```

```python
# Compreensão de listas - Modificando valores versão 1
numeros = [1, 30, 21, 2, 9, 65, 34]
quadrado = []

print(f"Lista números = {numeros}")
print(f"Lista quadrado dos números = {quadrado}\n")

for numero in numeros:
    quadrado.append(numero ** 2)

print(f"Lista quadrado dos números populada = {quadrado}")
```

```python
# Compreensão de listas - Modificando valores versão 2
numeros = [1, 30, 21, 2, 9, 65, 34]
quadrado = []

print(f"Lista números = {numeros}")
print(f"Lista quadrado dos números = {quadrado}\n")

# retorno obrigatório (quadrado do número), iteração for,
quadrado = [(numero ** 2) for numero in numeros]
print(f"Lista quadrado dos números populada = {quadrado}")
```

## 1.8. Tuplas em Python
Tuplas são estruturas de dados muito parecidas com as Listas, a principal diferença é que Tuplas são ***imutáveis*** enquanto Listas são ***mutáveis***.  Pordemos criar tuplas através da classe **tuple**, ou colocando valores separados por vírgula e parênteses.

```python
# Exemplos
frutas = ("laranja", "pera", "uva", "banana")
print(f"frutas = {frutas}\n")

letras = tuple("Python")
print(f"tuple('Python') = {letras}\n")

numeros = tuple([1, 2, 3, 4])
print(f"tuple([1, 2, 3, 4] = {numeros}\n")

pais = ("Brasil",)
print(f"pais = {pais}")
```

### 1.8.1. Acesso direto
A tupla é uma sequência, portanto podemos acessar seus dados utilizando índices.  Contamos o índice de determinada sequência a partir do zero.

```python
# Acesso direto
frutas = ("laranja", "pera", "uva", "banana")
print(f"frutas = {frutas}\n")
print(f"frutas[0] = {frutas[0]}\n")
print(f"frutas[3] = {frutas[3]}\n")
print(f"frutas[-1] = {frutas[-1]}\n")
print(f"frutas[-2] = {frutas[-2]}")
```

### 1.8.2. Tuplas aninhadas
Tuplas podem armazenar todos os tipos de objetos Python, portanto podemos ter tuplas que armazenam outras tuplas.  Com isso podemos criar estruturas bidimensionais (tabelas), e acessar informando os índices de linha e coluna, assim como nas Listas.

```python
# Tuplas Aninhadas
matriz = (
    ("a", 1, 2),
    (3, "b", 4),
    (5, 6, "c")
)
print(f"matriz = {matriz}\f")

print(f"matriz[0] = {matriz[0]}\f")
print(f"matriz[0][0] = {matriz[0][0]}\f")
print(f"matriz[0][-1] = {matriz[0][-1]}\f")
print(f"matriz[-1][-1] = {matriz[-1][-1]}")
```

### 1.8.3. Fatiamento de Tuplas
Além de acessar elementos diretamente, podemos extrair um conjunto de valores de uma sequência.  Para isso basta passar o índice inicial e/ou final para acessar o conjunto.  Podemos ainda informar quantas posições o cursor deve "pular" no acesso.

```python
# Fatiamento de Tuplas
tupla = tuple("Python")
print(f"tupla = {tupla}\n")

print(f"tupla[2:] = {tupla[2:]}")
print(f"tupla[:2] = {tupla[:2]}")
print(f"tupla[1:3] = {tupla[1:3]}")
print(f"tupla[0:3:2] = {tupla[0:3:2]}")
print(f"tupla[::-1] = {tupla[::-1]}")
```

### 1.8.4. Iterar Tuplas
A forma mais comum para percorrer os dados de uma Tupla é utilizando o comando **for**.

```python
# Iterando Tuplas com for
carros = ("gol", "celta", "palio", "fiesta")
print(f"carros = {carros}\n")

for indice, carro in enumerate(carros):
    print(f"carros[{indice}] = {carro}")
print("")

for carro in carros:
    print(carro)
```

## 1.9. Conjuntos (Sets)
Um set é uma coleção que não possui objetos repetidos, usamos sets para representar conjuntos matemáticos ou eliminar itens duplicados de um iterável.

```python
# Exemplos
exemplo1 = set([1, 2, 3, 1, 3, 4])
print(f"set([1, 2, 3, 1, 3, 4]) = {exemplo1}\n")

exemplo2 = set("abacaxi")
print(f"set('abacaxi') = {exemplo2}\n")

exemplo3 = set(("palio", "gol", "celta", "fiesta", "palio"))
print(f"set(('palio', 'gol', 'celta', 'fiesta', 'palio')) = {exemplo3}\n")

linguagens = {"Python", "Java", "Python", "Scala"}
print(f"linguagens = {"Python", "Java", "Python", "Scala"} = {linguagens}\n")
```

### 1.9.1. Acessando os Dados
Conjuntos em Python não suportam indexação e nem fatiamento, caso queira acessar os seus valores é necessário converter o conjunto para lista.

```python
# Exemplos
numeros = {1, 2, 3, 2}
print(f"numeros = {1, 2, 3, 2} = {numeros}\n")

numeros = list(numeros) # conversão de set para list
print(f"numeros[0] = {numeros[0]}\n")

# Iterando via for
carros = {"palio", "gol", "celta", "fiesta"}
print(f"carros = {carros}")

for carro in carros:
    print(carro)
```

### 1.9.2. Função enumerate
Às vezes é necessário saber qual o índice do objeto dentro do laço **for** num set.  Para isso, podemos usar a função **enumerate**.

```python
# Enumerate
carros = {"palio", "gol", "celta", "fiesta"}
print(f"carros = {carros}")

for indice, carro in enumerate(carros):
    print(f"indice {indice}: {carro}")
```

## 1.10. Dicionários
Um dicionário é um conjunto não-ordenado de pares chave:valor, onde as chaves são únicas em uma dada instância do dicionário.  Dicionários são delimitados por chaves: {}, e contém uma lista de pares chave:valor separada por vírgulas.

```python
# Exemplos
pessoa = {"nome": "Guilherme", "idade": 28}
print(f"Dicionário pessoa = {pessoa}\n")

gente = dict(nome = "Guilherme", idade = 28)
print(f"Dicionário gente = {gente}\n")

pessoa["telefone"] = "3333-1234"
print(f"Dicionário pessoa = {pessoa}")
```

### 1.10.1. Acessando os dados do Dicionário

```python
# Exemplos
dados = {"nome": "Guilherme", "idade": 28, "telefone": "3333-1234"}
print(f"dados = {dados}\n")

print(f"dados['nome'] = {dados["nome"]}")
print(f"dados['idade'] = {dados["idade"]}")
print(f"dados['telefone'] = {dados["telefone"]}\n")

dados["nome"] = "Maria" # substituindo dados
dados["idade"] = 18 # substituindo dados
dados["telefone"] = "9988-1781" # substituindo dados
print(f"dados = {dados}")
```

## 1.11. Funções
Função é um bloco de código identificado por um nome e pode receber uma lista de parâmetros, esses parâmetros podem ou não ter valores padrões.  Usar funções torna o código mais legível e possibilita o reaproveitamento de código.  Programar baseado em funções, é o mesmo que quizer que estamos programando de maneira estruturada.

```python
# Exemplos
def exibir_mensagem():
    print("Olá mundo!\n")

def exibir_mensagem_2(nome):
    print(f"Seja bem-vindo {nome}!")

def exibir_mensagem_3(nome = "Anônimo"):
    print(f"Seja bem-vindo {nome}!")

exibir_mensagem()
exibir_mensagem_2(nome = input("Digite o seu nome :"))
exibir_mensagem_3()
exibir_mensagem_3(nome = "Vinícius")
```

### 1.11.1. Retornando valores
Para retornar um valor, utilizamos a palavra reservada ***return***.  Toda função Python retorna **None** por padrão.  Diferente de outras linguagens de programação, em Python uma função pode retornar mais de um valor.

```python
# Exemplos
def calcular_total(numeros):
    return sum(numeros) # retorna a soma dos números da lista

def retorna_antecessor_e_sucessor(numero):
    antecessor = numero - 1
    sucessor = numero + 1
    return antecessor, sucessor # retorna uma tupla (antecessor, sucessor)

def funcao():
    print("Retorno de uma função sem retorno:")

def funcao2():
    print("Forçando o retorno de None:")
    return None # retorna None (não seria necessário, visto que é o padrão)

print(f"calcular_total([10, 20, 34]) = {calcular_total([10, 20, 34])}\n")
print(f"retorna_antecessor_e_sucessor(10) = {retorna_antecessor_e_sucessor(10)}\n")
print(f"{funcao()}\n")
print(funcao2())
```

### 1.11.2. Argumentos nomeados
Funções também podem ser chamadas usando argumentos nomeados na forma ***chave = valor***.

```python
# Exemplos
def salvar_carro(marca, modelo, ano, placa):
    """
    Argumentos = marca, modelo, ano e placa de um carro
    Imprime os argumentos
    Retorna: None
    """
    print(f"Carro inserido com sucesso: {marca} / {modelo} / {ano} / {placa}\n")

salvar_carro("Fiat", "Palio", 1999, "ABC-1234") # passando somente os valores, na ordem correta dos argumentos da função (posicionado)
salvar_carro(marca = "Fiat", modelo = "Palio", ano = 1999, placa = "ABC-1234") # passando os valores com suas chaves, na mesma ordem da função (nomeado)
salvar_carro(modelo = "Palio", ano = 1999, marca = "Fiat", placa = "ABC-1234") # passando os valores com suas chaves, fora de ordem da função (nomeado)
salvar_carro(**{"marca": "Fiat", "modelo": "Palio", "ano": 1999, "placa": "ABC-1234"}) # passando um dicionário para a função (nomeado)
```

#### 1.11.2.1 Args e kwargs
Podemos combinar parâmetros obrigatórios com args e kwargs.  Quando esses são definidos (*args e **kwargs), o método recebe os valores como tupla e dicionário respectivamente.

```python
# Exemplo
def exibir_poema(data_extenso, *args, **kwargs):
    texto = "\n".join(args)
    meta_dados = "\n".join([f"{chave.title()}: {valor}" for chave, valor in kwargs.items()])
    mensagem = f"{data_extenso}\n\n{texto}\n\n{meta_dados}\n"
    print(mensagem)

exibir_poema("Sábado, 6 de dezembro de 2025", # data_extenso
             "Zen of Python", # início *args
             "Beautiful is better than ugly.", 
             "Expliciti is better than implicit.",
             "Simple is better than complex.",
             "Complex é better than complicated.", # término *args
             autor = "Tim Peters", # início **kwargs
             ano = 1999) # término **kwargs

def exibir_poema2(data_extenso, *lista, **dicionario):
    texto = "\n".join(lista)
    meta_dados = "\n".join([f"{chave.title()}: {valor}" for chave, valor in dicionario.items()])
    mensagem = f"{data_extenso}\n\n{texto}\n\n{meta_dados}"
    print(mensagem)

exibir_poema2("Sábado, 6 de dezembro de 2025", # data_extenso
             "Zen of Python", # início *lista
             "Beautiful is better than ugly.", 
             "Expliciti is better than implicit.",
             "Simple is better than complex.",
             "Complex é better than complicated.", # término *lista
             autor = "Tim Peters", # início **dicionario
             ano = 1999) # término **dicionario
```

### 1.11.4. Objetos de Primeira Classe
Em Python tudo é objeto, dessa forma **funções também são objetos** o que as tornam objetos de primeira classe. Com isso podemos **atribuir funções a variáveis, passá-las como parâmetro para funções, usá-las como valores em estruturas de dados** (listas, tuplas, dicionários, etc) e usar como valor de retorno para uma função (closures).

```python
# Exemplo
def mais(a, b):
    return a + b

def menos(a, b):
    return a - b

def vezes(a, b):
    return a * b

def dividido(a, b):
    return a / b

def elevado(a, b):
    return a ** b

def exibir_resultado(a, b, funcao):
    resultado = funcao(a, b)
    print(f"O resultado da operação {a} {funcao.__name__} {b} = {resultado}\n")

exibir_resultado(20, 10, mais)
exibir_resultado(20, 10, menos)
exibir_resultado(20, 10, vezes)
exibir_resultado(20, 10, dividido)
exibir_resultado(2, 3, elevado)

operacao = mais
print(f"10 + 5 = {operacao(10, 5)}")

operacao = menos
print(f"10 - 5 = {operacao(10, 5)}")
```

### 1.11.5. Escopo Global e Escopo Local
Python trabalha com escopo local e global, dentro do bloco da função o escopo é local. Portanto alterações ali feitas em objetos imutáveis serão perdidas quando o método terminar de ser executado. Para usar objetos globais utilizamos a palavra-chave ***global***, que informa ao interpretador que a variável que está sendo manipulada no escopo local é global. Essa **NÃO é uma boa prática e deve ser evitada**.

```python
# Exemplo
salario = 2000
lista = [1]

def salario_bonus(bonus, lista):
    global salario
    salario += bonus
    lista = lista.append(2)
    return salario

print(f"Salário ({salario}) + bonus (500) = {salario_bonus(500, lista)}\n")
print(f"Lista = {lista}") # modificou, pois foi passada como argumento
```

### 1.12.1 Desafio 1
Faça um programa que calcule e imprima o salário a ser transferido para um funcionário.
Para realizar o cálculo, receba o valor bruto do salário e o adicional de benefícios.
O Salário a ser transferido é calculado da seguinte maneira:

(valor bruto do salário - percentual de imposto mediante ao salário) + adicional de benefícios

Para calcular o percentual de imposto, segue as alíquotas abaixo:
- De R$ 0.00 a 1100.00 (inclusive): 5% de imposto;

- De R$ 1100.01 a 2500.00 (inclusive): 10% de imposto;

- Acima de R$ 2500.00: 15% de imposto.

Entrada

A entrada consiste em vários arquivos de teste, que conterá o valor bruto do salário e adicional
dos benefícios.

Saída

Para cada arquivo da entrada, terá um arquivo de saída. E como mencionado no Desafio, será gerado uma
linha com um número que será a diferença entre o valor bruto do salário e o percentual de imposto mediante
a faixa salarial somado com o adicional de benefícios. Use o exemplo abaixo para clarear o que está sendo pedido.

Exemplo

Entradas = 2.000.00 e 250, Saída = 2050.00

Para ler e escrever os dados em Python, utilizamos as seguintes funções:
- input(): Lê uma linha com dado(s) de Entrada (Inputs) do usuário;
- print(): Imprime um texto de Saída (Output) e pula uma linha.

```python
# Desafio 1
# Função para calcular o imposto (baseado nas alíquotas fornecidas)
def calcular_imposto(salario):
    aliquota = 0.00
    if (salario >= 0.0 and salario <= 1100.0):
        aliquota = 0.05
    elif (salario > 1100.0 and salario <= 2500.0):
        aliquota = 0.10
    else:
        aliquota = 0.15
    return salario * aliquota

# Leitura dos valores de Entrada
valor_salario = float(input("Digite o valor do Salário Bruto : "))
valor_beneficios = float(input("Digite o valor dos Benefícios : "))

# Cálculo do imposto
valor_imposto = calcular_imposto(valor_salario)

# Cálculo do salário a ser transferido
salario_transferido = (valor_salario - valor_imposto) + valor_beneficios

# Impressão do salário a ser transferido
print(f"Salário Líquido a ser transferido : R$ {salario_transferido:.2f}")
```
